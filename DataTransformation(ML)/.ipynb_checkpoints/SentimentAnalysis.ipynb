```python
data = pd.read_csv('tsla_tweets.csv')

data.head()
```

```python
def cleanText(text):

  text = re.sub(r'@[A-Za-z0-9]+', '', text) #Removes @mentions
  text = re.sub(r'#', '', text) #Removes the # symbol
  text = re.sub(r'RT[\s]+', '', text) #Removes RT
  text = re.sub(r'https?:\/\/\S+','', text) #Removes hyperlinks
  #not cleaning emojis as the vader sentiment analysis takes those into account 

  return text

#Clean text and return
data['Tweet'] = data['Tweet'].astype(str) #Change the tweet data type from object to string
data['Tweet'] = data['Tweet'].apply(cleanText)
data
```

```python
# Assigns polarity for each tweet and gives 4 scores: negative, neutral, positive and compound(normalizing the other 3 scores)
# Compound score: between -1(most extreme negative) and +1 (most extreme positive)
# overall positive : (compound score >= 0.05) 
# overall neutral : (compound score > -0.05) and (compound score < 0.05) 
# overall negative : (compound score <= -0.05)
sia = SentimentIntensityAnalyzer()
data['Sentiments'] = data['Tweet'].apply(lambda Tweet: sia.polarity_scores(Tweet))
data = pd.concat([data.drop(['Sentiments'], axis=1), data['Sentiments'].apply(pd.Series)], axis=1)

data
```

```python
# Show percentage of positive, neutral and negative tweets 
positive = 0
neutral = 0
negative = 0
for i in range (0, len(data)):
    get_val = data.compound[i]
    if(float(get_val) <= -0.05):
        negative = negative + 1
    elif(float(get_val) >= 0.05):
        positive = positive + 1
    else:
        neutral = neutral + 1
posperc = (positive/(len(data)))*100
neuperc = (neutral/(len(data)))*100
negperc = (negative/(len(data)))*100
print("% of positive tweets= ", posperc)
print("% of neutral tweets= ", neuperc)
print("% of negative tweets= ", negperc)
arr = np.asarray([posperc, neuperc, negperc], dtype=int)
plt.pie(arr,labels=['positive','neutral', 'negative'])
plt.plot()
```

```python
# Daily average sentiment score
totalScore = 0
for i in range (0, len(data)):
  totalScore = totalScore + data.compound[i]
avgScore = totalScore/len(data)

avgScore
```
